```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
df = pd.read_csv('bank-data.csv')
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
df.drop('id', inplace = True, axis = 1)
df.drop('current_act', inplace = True, axis = 1)
df.income = le.fit_transform(df.income)
df.age = le.fit_transform(df.age)
df.sex = le.fit_transform(df.sex)
df.region = le.fit_transform(df.region)
df.married = le.fit_transform(df.married)
df.children = le.fit_transform(df.children)
df.car = le.fit_transform(df.car)
df.save_act = le.fit_transform(df.save_act)
# df.current_act = le.fit_transform(df.current_act)
df.mortgage = le.fit_transform(df.mortgage)
df.pep = le.fit_transform(df.pep)
```

```python
df.head()
```

```python
min_max_scalar = preprocessing.MinMaxScaler()
d = min_max_scalar.fit_transform(df)
c = ['age', 'sex', 'region', 'income', 'married', 'children', 'car', 'save_act', 'mortgage']
for column in c:
    df[column] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())
```

```python
X = df[df.columns [: - 1]]
```

```python
Y = df['pep']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.20, random_state = 30)
```

```python
gnb = GaussianNB()
```

```python
Y_pred = gnb.fit(X_train, Y_train).predict(X_test)
```

```python
print(accuracy_score(Y_test, Y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# After removing current_act

knn = KNeighborsClassifier(n_neighbors = 3, metric = 'euclidean')
knn.fit(X_train, Y_train)

predictions = knn.predict(X_test)

print(classification_report(Y_test, predictions))
print("Confusion Matrix")
print(confusion_matrix(Y_test, predictions))
print("\nAccuracy")
print(accuracy_score(Y_test, predictions))
```
